## keypoint 데이터 로드

```python
import os
import pandas as pd
import numpy as np
```

```python
npy_directory = "./npy/train"
f_list = os.listdir(npy_directory)
f_list.sort()
npy_files = [np.load(os.path.join(npy_directory, f)) for f in f_list if f.endswith(".npy")]

train_data = np.array(npy_files)
```

```python
f_list
```

```python
# df에 기본적으로 keypoint_0부터 keypoint_32까지 있어야함. 각 feature는 x, y, z 좌표를 가짐
df_columns = []
for i in range(33):
    df_columns.append("keypoint_" + str(i))

df_data = []
for data in train_data:
    df_data.append([np.array([d[0], d[1], d[2]]) for d in data])

df_train = pd.DataFrame(df_data, columns=df_columns)
df_train
```

### 엉덩이 중심으로 다시 좌표 설정

- 23(right hip)을 기준으로 좌표를 다시 설정한다.

```python
data_array = df_train.values

# Get the hip coordinates (assuming it's in columns 23)
hip_coords = np.array(data_array[:, 23])

# Subtract hip coordinates from all other coordinates
for i in range(data_array.shape[1]):
    data_array[:, i] -= hip_coords

# Now replace the original DataFrame with the new values
df_train = pd.DataFrame(data_array, columns=df_train.columns)
# df_train의 23번째(right hip coordinates)
df_train.iloc[:, 23]
```

```python
df_train
```

### 관절 각도

```python
import numpy as np

def calculate_angle(a, b, c):
    ba = a - b # vector from point b to a
    bc = c - b # vector from point b to c

    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = np.arccos(cosine_angle)

    return np.degrees(angle)
```

```python
for idx, data in enumerate(df_train.values):
    shoulder_right = data[11]
    elbow_right = data[13]
    wrist_right = data[15]

    # Left arm
    shoulder_left = data[12]
    elbow_left = data[14]
    wrist_left= data[16]
    

    angle_right_arm= calculate_angle(shoulder_right, elbow_right, wrist_right)
    angle_left_arm= calculate_angle(shoulder_left, elbow_left, wrist_left)

    print("Index: ", idx)
    print("Right arm angle: ", angle_right_arm)
    print("Left arm angle: ", angle_left_arm)

    # 소수점 2자리까지만 df의 현재 행에 저장
    df_train.loc[idx, "angle_right_arm"] = round(angle_right_arm, 2)
    df_train.loc[idx, "angle_left_arm"] = round(angle_left_arm, 2)
```

```python
df_train.index = f_list
df_train.to_csv("./230922_18h16m_train.csv", index=False)
df_train
```

## 정규화

```python
from sklearn.preprocessing import MinMaxScaler

# 1. Split each coordinate into individual columns
df_train_split = df_train.apply(lambda x: pd.Series([i for coord in x for i in coord]), axis=1)

# 2. Apply MinMaxScaler
scaler = MinMaxScaler()
scaled_train_data = scaler.fit_transform(df_train_split)

scaled_train_data.shape, scaled_train_data
```

```python
for idx, data in enumerate(df_train.values):
    shoulder_right = data[11]
    elbow_right = data[13]
    wrist_right = data[15]

    # Left arm
    shoulder_left = data[12]
    elbow_left = data[14]
    wrist_left= data[16]
    

    angle_right_arm= calculate_angle(shoulder_right, elbow_right, wrist_right)
    angle_left_arm= calculate_angle(shoulder_left, elbow_left, wrist_left)

    print("Index: ", idx)
    print("Right arm angle: ", angle_right_arm)
    print("Left arm angle: ", angle_left_arm)

    # 소수점 2자리까지만 df의 현재 행에 저장
    df_train.loc[idx, "angle_right_arm"] = round(angle_right_arm, 2)
    df_train.loc[idx, "angle_left_arm"] = round(angle_left_arm, 2)
```

```python
reshaped_scaled_data = scaled_train_data.reshape((210, 35, 3))

df_columns = []
for i in range(33):
    df_columns.append("keypoint_" + str(i))

df_columns.append("angle_right_arm")
df_columns.append("angle_left_arm")

# Convert each coordinate back to a list and store in a DataFrame
df_train_scaled = pd.DataFrame([[list(coord) for coord in row] for row in reshaped_scaled_data], columns=df_columns)
df_train_scaled
```
